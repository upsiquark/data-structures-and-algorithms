```python
#A code implementation of partitioning

class SortableArray:
    attr_reader = []

    def __init__(self, array):
        self.array = array

    def partition(self, left_pointer, right_pointer):

        #The right-most element is the pivot
        #Save the index of the pivot for later
        pivot_index = right_pointer

        #Save the pivot value
        pivot = self.array[pivot_index]

        #The right pointer begins one value left of the pivot
        right_pointer -= 1

        while True:

            #Move the left pointer to the right
            #so long as the number it currently points to is less than the pivot
            while self.array[left_pointer] < pivot:
                left_pointer += 1

            #Move the right pointer to the left
            #so long as the number it currently points to is greater than the pivot
            while self.array[right_pointer] > pivot:
                right_pointer -= 1

            #We have concluded the shifting of both pointers
            #so we check if the left pointer has moved to or beyond
            #the right pointer.

            #If it has, then we break the loop and prepare to swap the pivot
            if (left_pointer >= right_pointer):
                break

            #If the left pointer is remains left of the right pointer
            #then we swap the values at the left and right pointers
            else:
                self.array[left_pointer], self.array[right_pointer] = self.array[right_pointer], self.array[left_pointer]

                #Move the left pointer to the right
                #prepare for the next phase of pointer shifts
                left_pointer += 1

        #For the final step of partitioning, swap the value
        #of the left pointer with the pivot
        self.array[left_pointer], self.array[pivot_index] = self.array[pivot_index], self.array[left_pointer]

        #Then end the class by returning the left pointer in anticipation of the quicksort method
        return left_pointer
```

```python
#Now, let's implement quicksort
#A code implementation of partitioning

class SortableArray:
    attr_reader = []

    def __init__(self, array):
        self.array = array

    def partition(self, left_pointer, right_pointer):

        #The right-most element is the pivot
        #Save the index of the pivot for later
        pivot_index = right_pointer

        #Save the pivot value
        pivot = self.array[pivot_index]

        #The right pointer begins one value left of the pivot
        right_pointer -= 1

        while True:

            #Move the left pointer to the right
            #so long as the number it currently points to is less than the pivot
            while self.array[left_pointer] < pivot:
                left_pointer += 1

            #Move the right pointer to the left
            #so long as the number it currently points to is greater than the pivot
            while self.array[right_pointer] > pivot:
                right_pointer -= 1

            #We have concluded the shifting of both pointers
            #so we check if the left pointer has moved to or beyond
            #the right pointer.

            #If it has, then we break the loop and prepare to swap the pivot
            if (left_pointer >= right_pointer):
                break

            #If the left pointer is remains left of the right pointer
            #then we swap the values at the left and right pointers
            else:
                self.array[left_pointer], self.array[right_pointer] = self.array[right_pointer], self.array[left_pointer]

                #Move the left pointer to the right
                #prepare for the next phase of pointer shifts
                left_pointer += 1

        #For the final step of partitioning, swap the value
        #of the left pointer with the pivot
        self.array[left_pointer], self.array[pivot_index] = self.array[pivot_index], self.array[left_pointer]

        #Then end the class by returning the left pointer in anticipation of the quicksort method
        return left_pointer

    def quicksort(self, left_index, right_index):
        #Our base case is the subarray containing 0 or 1 elements
        if (right_index - left_index <= 0):
            return

        #Partition range of elements, save index of pivot
        pivot_index = self.partition(left_index, right_index)

        #Call the quicksort method recursively:
        
        #on anything left of the pivot
        self.quicksort(left_index, pivot_index - 1)

        #on anything right of the pivot
        self.quicksort(pivot_index + 1, right_index)
```
