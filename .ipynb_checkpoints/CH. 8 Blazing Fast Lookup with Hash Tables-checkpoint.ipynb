```python
#Here we have a function that returns the meaning of common HTTP status code numbers

def status_code_meaning(number):
    if (number == 200):
        return "OK"
    elif (number == 301):
        return "Moved Permanently"
    elif (number == 401):
        return "Unauthorized"
    elif (number == 404):
        return "Not Found"
    elif (number == 500):
        return "Internal Server Error"
```

```python
#We can use a hash table instead, and rid ourselves of the conditional logic

STATUS_CODES = { 200 : "OK", 303 : "Moved Permanently", 401 : "Unauthorized", 404 : "Not Found", 500 : "Internal Server Error"}

def status_code_meaning_1(number):
    return STATUS_CODES[number]
```

```python
#Let's write a function that compares two arrays and lets us know
#if one is a subset of the other

def isSubset(array1, array2):
    largerArray = []
    smallerArray = []
    
    #find out which array is smaller
    if (len(array1) > len(array2)):
        largerArray = array1
        smallerArray = array2
    else:
        largerArray = array2
        smallerArray = array1
        
    #iterate through smaller array
    for i in range(len(smallerArray)):
        
        #Assume that the current value in smaller array
        #does not appear in the larger array
        foundMatch = False
        
        #For each value in the smaller array, 
        #parse through the larger array

        for j in range(len(largerArray)):

            #If the two values are equivalent, then the current value
            #in the smaller array appears in the larger array:
            if (smallerArray[i] == largerArray[j]):
                foundMatch = True
                break

        #If the current value in the smaller array does not exist,
        #in the larger array, return False:
        if (foundMatch == False):
            return False
    #If we finish every loop,
    #then every value from the smaller array is present
    #in the larger array
    return True
```

```python
#Now lets use hash tables for the same problem!

def isSubset1(array1, array2):
    largerArray = []
    smallerArray = []
    hashTable = {}
    
    #Determine the which array is smaller
    if (len(array1) > len(array2)):
        largerArray = array1
        smallerArray = array2
    else:
        largerArray = array2
        smallerArray = array1
        
    #store all elements of the larger array in hashTable:
    for i in largerArray:
        hashTable[i] = True

    #iterate through each element in smallerArray,
    #then return False if we run into an element not in hashTable 
    for j in smallerArray:
        if ((j in hashTable) == False):
            return False

    #if the function makes it this far without returning False,
    #then all the items in smallerArray
    #were in largerArray
    return True
```

```python
#Write a function that returns the intersection of two arrays.

def intersectionTwo(array1, array2):
    largerArray = []
    smallerArray = []
    intersectArray = []
    
    hashTable = {}
    
    if (len(array1) > len(array2)):
        largerArray = array1
        smallerArray = array2
    else:
        largerArray = array2
        smallerArray = array1
        
    #store all elements of the larger array in hashTable:
    for i in largerArray:
        hashTable[i] = True

    #iterate through each element in smallerArray,
    #then return False if we run into an element not in hashTable 
    for j in smallerArray:
        if ((j in hashTable)):
            intersectArray.append(j)
    #if the function makes it this far without returning False,
    #then all the items in smallerArray
    #were in largerArray
    return intersectArray
```

```python
#Write a function that accepts an array of strings and returns
#the first duplicate value it finds

def firstDuplicate(array):
    hashTable = {}

    for i in array:
        if (i in hashTable):
            return i
        hashTable[i] = True
```

```python
#Write a function that accepts a string that contains
#all the letters of the alphabet except one
#and returns the missing letter

def missingLetter(string):
    #There certainly is an easier way to generate the alphabet array,
    #but I wanted to refrain from outside help for these attempts
    
    alphabet = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"]
    hashTable = {}

    for i in string:
        hashTable[i] = True

    for j in alphabet:
        if ((j in hashTable) == False):
            return j
```

```python
#Write a function that returns the first non-duplicated character in a string

def firstnonDuplicate(string):
    hashTable = {}

    for i in string:
        if (i in hashTable):
            hashTable[i] += 1
        else:
            hashTable[i] = 0

    for j in hashTable:
        if hashTable[j] == 0:
            return j
```
